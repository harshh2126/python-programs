```python
# get size of tuple ,list and set
from sys import getsizeof
tuple_=(1,2,3,4)
print("Tuple: ",getsizeof(tuple_))
list_=[1,2,3,4]
print("List: ",getsizeof(list_))
size_of_set=set([1,2])
print("set:",getsizeof(size_of_set))
```

```python
# create even no's only, (using range)
set_of_even =set(range(1,11))
#using set() function
empty_set=set()
for i in set_of_even:
    if i%2==0:
        empty_set.add(i)
print(empty_set)

```

```python
# get size of set (set allows only immutable data types[such as tuple]
#list is also allowed but not nested list of elements ex:- set([1,2,3,[5,6],7])
#dict is also not allowed in set()
from sys import getsizeof
size_of_set=set([1,2,3,4,5,6,'a'])
print("set:",getsizeof(size_of_set))
```

```python
# separating even and odd numbers from 1 to 100 using list in python
even=[]
odd=[]
list_num=list(range(1,101))
for num in list_num:
    if num%2==0:
        even.append(num)
    else:
        odd.append(num)
print("the even numbers are:",even,"\n")
print("the odd numbers are:",odd)
```

```python
#take one lists of duplicate numbers and make dictionary which key is list element 
#and value is number of occurrences of that element in that list( using count() method )
list_with_dupli=[1,2,3,1,4,3,4,2,]
#create empty list with duplicates
dictt_={}
#create empty dict
for i in list_with_dupli:
    dictt_[i]=list_with_dupli.count(i)
#using count method to know no of occurances
```
